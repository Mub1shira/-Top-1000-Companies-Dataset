# <h1 style="color:orange;">Visualizations</h2>

<h1 style="color:green;">Correlation Analysis</h2> 

```python
# Select only the numerical columns for correlation analysis
numerical_columns = ['employees', 'founded', 'GrowjoRanking', 'Previous Ranking', 'estimated_revenues', 'job_openings',
                     'valuation', 'total_funding', 'growth_percentage']

# Calculate the correlation matrix
correlation_matrix = df[numerical_columns].corr()

# Create a heatmap of the correlation matrix
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

 <h1 style="color:green;">Histogram of Employee Count</h2>

```python
# Plot a histogram of employee counts
plt.figure(figsize=(10, 6))
sns.histplot(df['employees'], bins=30, kde=True)
plt.xlabel('Employee Count')
plt.ylabel('Frequency')
plt.title('Distribution of Employee Count')
plt.show()
```

  <h1 style="color:green;">Bar Chart of Top Industries</h2>

```python
# Count the number of companies in each industry
top_industries = df['Industry'].value_counts().head(10)

# Plot a bar chart of the top 10 industries
plt.figure(figsize=(12, 6))
sns.barplot(x=top_industries.index, y=top_industries.values)
plt.xticks(rotation=90)
plt.xlabel('Industry')
plt.ylabel('Number of Companies')
plt.title('Top 10 Industries by Number of Companies')
plt.show()
```

  <h1 style="color:green;">Scatter Plot of Estimated Revenues vs. Employee Count</h2>

```python
# Plot a scatter plot of estimated revenues vs. employee count
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='estimated_revenues', y='employees', hue='Industry', alpha=0.5)
plt.xlabel('Estimated Revenues')
plt.ylabel('Employee Count')
plt.title('Estimated Revenues vs. Employee Count by Industry')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

<h1 style="color:green;">Valuation vs. Total Funding</h2>

```python
# Interactive Scatter Plot of Valuation vs. Total Funding
fig = px.scatter(df, x='valuation', y='total_funding', color='Industry', hover_name='company_name')
fig.update_layout(title='Interactive Scatter Plot of Valuation vs. Total Funding')
fig.show()
```

<h1 style="color:green;">Distribution of Companies by Country</h2>

```python
# Drop rows with missing values in 'country' column
data_cleaned = df.dropna(subset=['country'])

# Create a DataFrame to store the counts of companies by country
country_counts = data_cleaned['country'].value_counts().reset_index()
country_counts.columns = ['Country', 'Number of Companies']

# Create a choropleth map using Plotly Express
fig = px.choropleth(
    country_counts, 
    locations='Country',
    locationmode='country names',
    color='Number of Companies',
    hover_name='Country',
    color_continuous_scale='Viridis',  # You can choose a different color scale
    title='Distribution of Companies by Country'
)

# Customize the map layout
fig.update_geos(
    visible=False, 
    showcoastlines=True,
    coastlinecolor="Black",
    projection_scale=7
)

# Show the map
fig.show()
```

```python
# Print unique values in the 'growth_percentage' column
unique_growth_percentage = df['growth_percentage'].unique()
print("Unique growth percentages:", unique_growth_percentage)

# Print unique values in the 'country' column
unique_countries = df['country'].unique()
print("Unique countries:", unique_countries)
```

```python
import re
# Extract numeric values from growth_percentage using regex and convert to numeric
df['growth_percentage'] = pd.to_numeric(df['growth_percentage'].str.replace(r'[^0-9.]', '', regex=True))
```

```python
# Average Growth Percentage by Country
avg_growth_by_country = df.groupby('country')['growth_percentage'].mean().sort_values(ascending=False)
plt.figure(figsize=(12, 6))
sns.barplot(x=avg_growth_by_country.index, y=avg_growth_by_country.values)
plt.title('Average Growth Percentage by Country')
plt.xlabel('Country')
plt.ylabel('Average Growth Percentage')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Box Plot of Growth Percentage by Country
plt.figure(figsize=(12, 6))
sns.boxplot(x='country', y='growth_percentage', data=df)
plt.title('Box Plot of Growth Percentage by Country')
plt.xlabel('country')
plt.ylabel('Growth Percentage')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility
plt.tight_layout()
plt.show()
```

```python
country_counts = df['country'].value_counts()

top_countries = country_counts.head(7)
other_countries_count = country_counts[7:].sum()
top_countries['Others'] = other_countries_count

plt.figure(figsize=(8, 8))
top_countries.plot(kind='pie', autopct='%1.1f%%')
plt.title('Country distribution')
plt.ylabel('')
plt.show()
```
